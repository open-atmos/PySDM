https://doi.org/10.5194/acp-16-2083-2016

```python
import os, sys
os.environ['NUMBA_THREADING_LAYER'] = 'workqueue'  # PySDM & PyMPDATA don't work with TBB; OpenMP has extra dependencies on macOS
if 'google.colab' in sys.modules:
    !pip --quiet install open-atmos-jupyter-utils
    from open_atmos_jupyter_utils import pip_install_on_colab
    pip_install_on_colab('PySDM-examples', 'PySDM')
```

```python
from open_atmos_jupyter_utils import show_plot
from PySDM_examples.Alpert_and_Knopf_2016 import Simulation, Table2
from matplotlib import pylab
from PySDM.physics import si
```

```python
cases = Table2()
```

```python
sim = Simulation(cases=cases, time_step=5*si.s, heterogeneous_ice_nucleation_rate='ABIFM',
                 temperature_range=(228*si.K, 258*si.K))
```

```python
sim.run(('Cr1', 'Cr2'))
```

```python
sim.plot_j_het(variant='apparent', ylim=(1e-4, 1e10), abifm_params_case='Cr1')
pylab.title("a")
show_plot('fig_4a.pdf')
```

```python
sim.plot_j_het(variant='actual', ylim=(1e-4, 1e10), abifm_params_case='Cr1')
pylab.title("b")
show_plot('fig_4b.pdf')
```

```python
sim.plot(ylim=(0, 1))
pylab.title("c, d")
show_plot('fig_4cd.pdf')
```
