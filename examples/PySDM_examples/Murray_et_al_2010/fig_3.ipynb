#### based on Fig. 3 from Murray et al. 2010 (Physical Chemistry Chemical Physics 12) "_Kinetics of the homogeneous freezing of water_" https://doi.org/10.1039/c003297b

```python
from open_atmos_jupyter_utils import show_plot
from PySDM_examples.Murray_et_al_2010 import Simulation
from PySDM.physics import si
```

```python
cases = { 'cr2.5': {'cooling_rate': 2.5 * si.K / si.minute, 'color':'purple'},
          'cr5': {'cooling_rate': 5 * si.K / si.minute, 'color':'green'},
          'cr7.5': {'cooling_rate': 7.5 * si.K / si.minute, 'color':'blue'},
          'cr10': {'cooling_rate': 10 * si.K / si.minute, 'color':'red'},
          }
```

```python
sim = Simulation(cases=cases, droplet_radius=10*si.micrometer, n_runs_per_case=5)
sim.run(('cr10','cr7.5','cr5','cr2.5'))
sim.plot_histogram()
show_plot('fig_3_10µm.pdf')
```

```python
sim_20 = Simulation(cases=cases, droplet_radius=20*si.micrometer)
sim_20.run(('cr10','cr7.5','cr5','cr2.5'))
sim_30 = Simulation(cases=cases, droplet_radius=30 * si.micrometer)
sim_30.run(('cr10','cr7.5','cr5','cr2.5'))
sim_40 = Simulation(cases=cases, droplet_radius=40 * si.micrometer)
sim_40.run(('cr10','cr7.5','cr5','cr2.5'))
```

```python
sim_20.plot_histogram()
show_plot('fig_3_20µm.pdf')
sim_30.plot_histogram()
show_plot('fig_3_30µm.pdf')
sim_40.plot_histogram()
show_plot('fig_3_40µm.pdf')
```
