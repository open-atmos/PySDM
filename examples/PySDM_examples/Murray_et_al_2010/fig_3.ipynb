#### based on Fig. 3 from Murray et al. 2010 (Physical Chemistry Chemical Physics 12) "_Kinetics of the homogeneous freezing of water_" https://doi.org/10.1039/c003297b

```python
from open_atmos_jupyter_utils import show_plot
from PySDM_examples.Murray_et_al_2010 import Simulation
from PySDM.physics import si
```

```python
cases = { 'cr2.5': {'cooling_rate': 2.5 * si.K / si.minute, 'color':'purple'},
          'cr5': {'cooling_rate': 5 * si.K / si.minute, 'color':'green'},
          'cr7.5': {'cooling_rate': 7.5 * si.K / si.minute, 'color':'blue'},
          'cr10': {'cooling_rate': 10 * si.K / si.minute, 'color':'red'},
          }
```

```python
sim = Simulation(cases=cases, time_step=0.1 * si.s, n_sd=1000, n_runs_per_case=1)
sim.run(('cr2.5','cr5','cr7.5','cr10'))
sim.plot_histogram()
show_plot('fig_3.pdf')
```
